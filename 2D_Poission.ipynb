```python
'''
Module to solve the Poisson equation with zero order term and Dirichlet boundary data

            - (u_xx + u_yy) + c*u  = f        in (0,L_x) x (0,L_y)
                        u(x,y)      = g       on space boundary 
                     
    with source source f rectangle (0,L_x) x (0,L_y) using central in space difference quotients for space discretization. 
'''
```

```python
# Import necessary modules
import numpy as np
import matplotlib.pyplot as plt

import scipy
import scipy.sparse as sp


from mpl_toolkits.mplot3d import Axes3D
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import spsolve
import poisson_solver
#from poisson_solver import FD_POISSON
```

```python
#------------------------------------------------------------------#
#                           PARAMETERS                             #
#------------------------------------------------------------------#
# DOMAIN AND DISCRETIZATION PARAMETERS
L_x = 1
L_y = 1
h = 0.05

#------------------------------------------------------------------#
#                 COEFFICIENT FUNCTIONS AND RHS                    #
#------------------------------------------------------------------#
# Set right hand side f
sig = 0.1
def f(x,y):
    return 100*np.exp((-(x-0.5) ** 2)/sig ** 2) * np.exp(-((y-0.5) ** 2)/sig ** 2) / (2*np.pi*sig)

# Set boundary condition g
def g(x,y):
    if y==0:
        return 1*np.sin(2*x*np.pi)
    elif y==L_y:
        return 1*np.sin(2*x*np.pi)
    elif x==0:
        return 2*np.sin(2*y*np.pi)
    elif x==L_x:
        return 2*np.sin(2*y*np.pi)

# Set coefficient function 
def c(x,y):
    return 0

```

```python
#------------------------------------------------------------------#
#                           SOLVE                                  #
#------------------------------------------------------------------#
# Create an instance of FD_Poisson
example1 = poisson_solver.FD_POISSON(L_x, L_y, h, c, f, g)

# Plot boundary conditions and solution
example1.plot_3D_boundary_conditions()
example1.plot_3D_solution()

```
